```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
%matplotlib inline

data = pd.read_csv("./wiki_movie_plots_deduped.csv")
data.head()
```

```python
data['Genre'].count()
```

```python
#get the proportion of each class in the data set
(data['Genre'].value_counts()/data['Genre'].count())*100
```

Clean the text:
● Remove stop words
● Remove numbers and other non-letter characters
● Stem the words

```python
#select the 5 classes we need
data1=data[data['Genre'].isin(['drama','comedy','adventure','romance','western'])]
data1
```

```python
#Plan to use 'Plot' to do the prediction and classification
df=data1[['Plot','Genre']]
df
```

```python
#Remove stop words
#Remove numbers and other non-letter characters
#Stem the words
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

stop_words=set(stopwords.words("english"))
remove=re.compile('[^a-zA-Z ]', re.UNICODE)
ps=PorterStemmer()

newplot=[]
for w in df['Plot']:
    m=remove.sub('', w)
    tokenized_word=word_tokenize(m)
    #newsent=[]
    newsent=""
    for a in tokenized_word:
        if a not in stop_words:
            #newsent.append(ps.stem(a))
            newsent=newsent+ps.stem(a)+" "
    newplot.append(newsent)
```

```python
newdf=pd.DataFrame({"Plot":newplot,"Genre":df['Genre']})
newdf
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect=CountVectorizer()
counts=count_vect.fit_transform(newdf['Plot'])
counts.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer=TfidfTransformer()
df_tdidf=tfidf_transformer.fit_transform(counts)
df_tdidf.shape
```

```python
df_tdidf.toarray()
```

Split the data randomly into training and testing set

```python
from sklearn.model_selection import train_test_split 

train, test = train_test_split(newdf,test_size = 0.3)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import confusion_matrix

c_vect=CountVectorizer()
c=c_vect.fit_transform(train.Plot)
c.shape
```

```python
t_transformer=TfidfTransformer()
X_train_t=t_transformer.fit_transform(c)
X_train_t.shape
```

Train SVM and report confusion matrix

```python
svm=Pipeline([('vect', CountVectorizer()),('tfidf', TfidfTransformer()),('clf-svm', SGDClassifier(loss='hinge', penalty='l2', alpha=1e-3,  random_state=42)),])
svm=svm.fit(train.Plot, train.Genre)

pre1=svm.predict(test.Plot)
accuracy_score(test.Genre, pre1)
```

```python
confusion_matrix(test.Genre, pre1)
```

Train Multinomial NB and report confusion matrix.

```python
clf=MultinomialNB().fit(X_train_t, train.Genre)
```

```python
nb=Pipeline([('vect', CountVectorizer()),('tfidf', TfidfTransformer()),('clf', MultinomialNB()),])
nb=nb.fit(train.Plot, train.Genre)

pre2=nb.predict(test.Plot)
accuracy_score(test.Genre, pre2)
```

```python
confusion_matrix(test.Genre, pre2)
```

SVM has the higher accuracy.
Reason:
Naive Bayes treats the data as independent, whereas SVM looks at the interactions between data to a certain degree. In this case, there are relationships between words. So SVM is better in this case.

```python
new_svm=Pipeline([('vect', CountVectorizer()),('tfidf', TfidfTransformer()),('clf-svm', SGDClassifier(loss='hinge', penalty='l2', alpha=1e-3,  random_state=100)),])
new_svm=new_svm.fit(train.Plot, train.Genre)

pre3=new_svm.predict(test.Plot)
accuracy_score(test.Genre, pre3)
```

```python
confusion_matrix(test.Genre, pre3)
```

```python
#extract the nouns
np1=[]
for w in df['Plot']:
    tokens=word_tokenize(w)
    posd=nltk.pos_tag(tokens)
    is_noun = lambda pos: pos[:2] == 'NN'
    nouns=[word for (word, pos) in posd if is_noun(pos)]
    nn=""
    for n in nouns:
        nn=nn+n+" "
    np1.append(nn)
```
